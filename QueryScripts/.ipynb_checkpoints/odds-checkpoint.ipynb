```python
import pandas as pd
import numpy as np

import urllib2
from bs4 import BeautifulSoup

pd.set_option('display.max_rows', 600)
```

```python
odds_data = {}

target_list = ['http://www.footballlocks.com/nfl_odds_week_1.shtml#Closing%20NFL%20Odds%20Week%201,%202015',
              'http://www.footballlocks.com/nfl_odds_week_2.shtml#Closing%20NFL%20Odds%20Week%202,%202015',
              'http://www.footballlocks.com/nfl_odds_week_3.shtml#Closing%20NFL%20Odds%20Week%203,%202015',
              'http://www.footballlocks.com/nfl_odds_week_4.shtml#Closing%20NFL%20Odds%20Week%204,%202015',
              'http://www.footballlocks.com/nfl_odds_week_5.shtml#Closing%20NFL%20Odds%20Week%205,%202015',
              'http://www.footballlocks.com/nfl_odds_week_6.shtml#Closing%20NFL%20Odds%20Week%206,%202015',
              'http://www.footballlocks.com/nfl_odds_week_7.shtml#Closing%20NFL%20Odds%20Week%207,%202015',
              'http://www.footballlocks.com/nfl_odds_week_8.shtml#Closing%20NFL%20Odds%20Week%208,%202015',
              'http://www.footballlocks.com/nfl_odds_week_9.shtml#Closing%20NFL%20Odds%20Week%209,%202015',
              'http://www.footballlocks.com/nfl_odds_week_10.shtml#Closing%20NFL%20Odds%20Week%2010,%202015',
              'http://www.footballlocks.com/nfl_odds_week_11.shtml#Closing%20NFL%20Odds%20Week%2011,%202015',
              'http://www.footballlocks.com/nfl_odds_week_12.shtml#Closing%20NFL%20Odds%20Week%2012,%202015',
              'http://www.footballlocks.com/nfl_odds_week_13.shtml#Closing%20NFL%20Odds%20Week%2013,%202015',
              'http://www.footballlocks.com/nfl_odds_week_14.shtml#Closing%20NFL%20Odds%20Week%2014,%202015',
              'http://www.footballlocks.com/nfl_odds_week_15.shtml#Closing%20NFL%20Odds%20Week%2015,%202015',
              'http://www.footballlocks.com/nfl_odds_week_16.shtml#Closing%20NFL%20Odds%20Week%2016,%202015',
              'http://www.footballlocks.com/nfl_odds_week_17.shtml#Closing%20NFL%20Odds%20Week%2017,%202016']

for target in enumerate(target_list, start=1):
    date = []
    favorite = []
    spread = []
    underdog = []
    total =[]
    moneyodds =[]
    
    page = urllib2.urlopen(target[1])
    

    soup = BeautifulSoup(page)
    week = soup.find('table', attrs={"width":"99%"})

    for i in week.findAll('table'):
        for cell in i.findAll('tr'):
            for data in enumerate(cell.findAll('td')):
                
                if len(data[1].contents)<2 and '<b>' not in str(data[1].contents) and '</td>' not in str(data[1].contents) and 'PPD' not in str(data[1].contents) or 'PK' in str(data[1].contents) and len(str(data[1].contents))==69 and '<b>' not in str(data[1].contents) and '</td>' not in str(data[1].contents) and 'PPD' not in str(data[1].contents) or len(data[1].contents)<=4 and ('London' or 'Toronto' in str(data[1].contents)) and '<b>' not in str(data[1].contents) and '<p' not in str(data[1].contents) and 'PPD' not in str(data[1].contents):
                    if str(data[1].contents)!= '[]':
                            if len(date)==len(favorite) and len(date)==len(spread) and len(date)==len(underdog) and len(date)==len(total) and len(date)==len(moneyodds):
                                if 'Buffalo' in data[1].contents[0] or 'Jets' in data[1].contents[0]:
                                    pass
                                else:
                                    date.append(data[1].contents[0])
                            elif len(favorite)==len(spread) and len(favorite)==len(underdog) and len(favorite)==len(total) and len(favorite)==len(moneyodds):
                                if 'London' in str(data[1].contents) or 'Toronto' in str(data[1].contents) or 'Detroit' in str(data[1].contents):
                                    favorite.append('At ' + data[1].contents[0].string.rstrip('\n').lstrip('At'))
                                else:
                                    favorite.append(data[1].contents[0])
                            elif len(spread)==len(underdog) and len(spread)==len(total) and len(spread)==len(moneyodds):
                                if len(data[1].contents)==2:
                                    spread.append('PK')
                                else:
                                    spread.append(data[1].contents[0].string)
                            elif len(underdog)==len(total) and len(underdog)==len(moneyodds):
                                if 'London' in str(data[1].contents) or 'Toronto' in str(data[1].contents):
                                    underdog.append('At ' + data[1].contents[0].string.rstrip('\n'))
                                else:
                                    underdog.append(data[1].contents[0])
                            elif len(total)==len(moneyodds):
                                total.append(data[1].contents[0])
                            else:
                                odds = data[1].text
                                n = odds.replace(u'\xa0', '')
                                moneyodds.append(n)


    datadict = {'date':date, 'favorite':favorite,  'spread':spread, 'underdog':underdog, 'total': total}
    frame = pd.DataFrame(data=datadict)
    frame = frame.drop_duplicates()
    odds_data['week_%s' % target[0]] = frame
    odds_data['week_%s' % target[0]].insert(5, 'week', target[0])
    odds_data['week_%s' % target[0]].insert(6, 'season', 2008)
    
    del datadict
    del frame
```

```python
db_games = q.game(season_type='Regular').as_games()

db_gsis = []
db_week = []
db_season = []
db_home_team = []
db_away_team = []

def convert_team(db_team=None):
    return teams['%s' % db_team]

for game in db_games:
    db_gsis.append(game.gsis_id)
    db_week.append(game.week)
    db_season.append(game.season_year)
    db_home_team.append(convert_team(game.home_team))
    db_away_team.append(convert_team(game.away_team))
    
db_data = {'gsis_id':db_gsis, 'week':db_week, 'season':db_season, 'home_team': db_home_team, 'away_team':db_away_team} 

db_frame = pd.DataFrame(data=db_data, index=db_data['gsis_id'])
db_frame
```
