```python
import amaricooper
import pandas as pd
import numpy as np


import urllib2

from bs4 import BeautifulSoup
```

```python
date = []
favorite = []
spread = []
underdog = []
total =[]
moneyodds =[]

pd.set_option('display.max_rows', 500)


target = "http://www.footballlocks.com/nfl_odds_week_1.shtml#Closing%20NFL%20Odds%20Week%201,%202015"
page = urllib2.urlopen(target)

soup = BeautifulSoup(page)
week1 = soup.find('table', attrs={"width":"99%"})#.find_all('table')

for i in week1.findAll('table'):
    for cell in i.findAll('tr'):
        for data in enumerate(cell.findAll('td')):
            if len(data[1].contents)<2 and '<b>' not in str(data[1].contents) and '</td>' not in str(data[1].contents):
                if str(data[1].contents)!= '[]':
                        #print len(date), len(favorite), len(spread), len(underdog), len(total), len(moneyodds)
                        if len(date)==len(favorite) and len(date)==len(spread) and len(date)==len(underdog) and len(date)==len(total) and len(date)==len(moneyodds):
                            date.append(data[1].contents[0])
                        elif len(favorite)==len(spread) and len(favorite)==len(underdog) and len(favorite)==len(total) and len(favorite)==len(moneyodds):
                            favorite.append(data[1].contents[0])
                        elif len(spread)==len(underdog) and len(spread)==len(total) and len(spread)==len(moneyodds):
                            spread.append(data[1].contents[0])
                        elif len(underdog)==len(total) and len(underdog)==len(moneyodds):
                            underdog.append(data[1].contents[0])
                        elif len(total)==len(moneyodds):
                            total.append(data[1].contents[0])
                        else:
                            odds = data[1].text
                            n = odds.replace(u'\xa0', '')
                            moneyodds.append(n)



#for i in week1.findAll('table'):
#    for table in i.findAll('table'):
#        row = table.findAll('tr')
#        for td in row:
#            if 'td' in str(td.contents[1]) and 'ET' in str(td.contents[1]):
#                print td.contents
                
   
#                date.append(td.contents[1].text)
#                favorite.append(td.contents[3].text)
#                spread.append(td.contents[5].text)
#                underdog.append(td.contents[7].text)
#                total.append(td.contents[9].text)
#                odds = unicode(td.contents[11])
#                moneyodds.append(odds.strip('td/<>\nxa'))
datadict = {'date':date, 'favorite':favorite,  'spread':spread, 'underdog':underdog, 'total': total}
odds_frame = pd.DataFrame(data=datadict)
odds_frame
```
